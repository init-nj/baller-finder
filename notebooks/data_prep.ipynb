```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
appearances = pd.read_csv('data/appearances.csv') #Mainly Stats
players = pd.read_csv('data/players.csv') #Player Info
country = pd.read_csv('data/national_teams.csv')
#Enough for Similar Player finder for now.
```

```python
print(appearances.columns)
print(players.columns)
print(country.columns)
```

```python
stats = appearances.merge(players[['player_id','country_of_birth', 'city_of_birth',
       'country_of_citizenship', 'date_of_birth', 'sub_position', 'position',
       'foot', 'height_in_cm', 'international_caps', 'international_goals',
       'current_national_team_id','market_value_in_eur', 'highest_market_value_in_eur']], on = 'player_id', how ='left')
```

```python
stats
```

```python
stats.rename(columns={'current_national_team_id':'country_id'}, inplace = True)
```

```python
stats = stats.merge(country[['country_id','fifa_ranking']], on='country_id', how='left')
```

```python
agg = stats.groupby('player_id').agg(
    total_goals = ('goals', 'sum'),
    total_assits = ('assists', 'sum'),
    total_minutes = ('minutes_played', 'sum'),
    total_appearances = ('game_id', 'count'),
    total_yellows = ('yellow_cards','sum'),
    total_reds = ('red_cards','sum')
).reset_index()
```

```python
agg['total_cards'] = agg['total_yellows']+agg['total_reds']
```

```python
appearances['total_cards'] = appearances['yellow_cards'] + appearances['red_cards']
```

```python
competition_avg = appearances.groupby('competition_id').agg(
    competition_avg_goals = ('goals', 'mean'),
    competition_avg_assists = ('assists', 'mean'),
    competition_avg_cards = ('total_cards', 'mean'),
    competition_avg_playtime = ('minutes_played', 'mean')
).reset_index()
```

```python
competition_avg
```

```python
def shrunk_rate(stat, minutes, league_avg_p90, weight=500):
    return (stat + league_avg_p90 * (weight / 90)) / ((minutes + weight) / 90)

# Compute league averages from all players
league_avg_goals   = (agg['total_goals'].sum()   / agg['total_minutes'].sum()) * 90
league_avg_assists = (agg['total_assits'].sum() / agg['total_minutes'].sum()) * 90

# Apply shrinkage — no player is dropped
agg['goals_p90']   = agg.apply(lambda r: shrunk_rate(r['total_goals'],   r['total_minutes'], league_avg_goals),   axis=1)
agg['assists_p90'] = agg.apply(lambda r: shrunk_rate(r['total_assits'], r['total_minutes'], league_avg_assists), axis=1)
```

```python
agg
```

```python
agg['low_confidence'] = agg['total_minutes'] < 200
```

```python
player_meta = players[['player_id', 'name', 'position','sub_position', 'date_of_birth']].drop_duplicates('player_id')

df = agg.merge(player_meta, on='player_id', how='left')

df['date_of_birth'] = pd.to_datetime(df['date_of_birth'], utc=True)
age_days = (pd.Timestamp('today', tz='UTC') - df['date_of_birth']).dt.days
df['age'] = (age_days / 365.25).apply(np.floor)
df['age'] = df['age'].astype('Int64')
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
df['position_encoded'] = LabelEncoder().fit_transform(df['sub_position'].fillna('Unknown'))
feature_cols = ['goals_p90', 'assists_p90', 'total_cards', 'age', 'position_encoded']
X_raw = df[feature_cols].fillna(0)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_raw)
```

```python
from sklearn.neighbors import NearestNeighbors

knn = NearestNeighbors(n_neighbors=6, metric='cosine')
knn.fit(X_scaled)
```

```python
def find_similar(player_name, n=5, same_position=True):
    matches = df[df['name'].str.contains(player_name, case=False, na=False)]
    if matches.empty:
        return f"Player '{player_name}' not found."
    
    idx = matches.index[0]
    distances, indices = knn.kneighbors([X_scaled[idx]], n_neighbors=n+1)
    
    similar = df.iloc[indices[0][1:]].copy()  # exclude the query player himself
    similar['similarity'] = 1 - distances[0][1:]
    
    if same_position:
        query_pos = df.iloc[idx]['sub_position']
        similar = similar[similar['sub_position'] == query_pos]
    
    return similar[['name', 'position','sub_position', 'age', 'goals_p90', 'assists_p90', 'similarity', 'low_confidence']]
```
